```python
for yr in range (st_year,ed_year+1):
    os.makedirs('D:/Rahma/GsMap/'+str(yr), exist_ok=True)
    print ('download...'+str(yr))
    for mt in range (1,13) :
        mti = format(mt,'02d')
        jday = monthrange(yr,mt)[1] 
        
        for dy in range (1,jday+1) :
            day = format(dy,'02d')
            
            filename = 'gsmap_nrt.'+str(yr)+mti+day+'.0.1d.daily.00Z-23Z.dat.gz'
            
            ftp = ftplib.FTP("hokusai.eorc.jaxa.jp")
            ftp.login("rainmap", "Niskur+1404")
            ftp.cwd('/realtime_ver/v6/daily0.1/00Z-23Z/'+str(yr)+mti)
            path='D:/Rahma/GsMap/'+str(yr)+'/'

            ftp.retrbinary("RETR " + filename ,open(os.path.join(path,filename), 'wb').write)
            
            print ('downloaded....'+filename)
        
            #print (jday)
            #print (filename)
```
