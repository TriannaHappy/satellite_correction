Pertama kita akan menentukan rumus IDW per stasiun hujan yang dianggap kosong

```python
# create a variable which contains of list of multiplier factor for each station
for i in range (1,len(dataset_rainfall.columns)):
    asal=dataset_rainfall.columns[i]
    print ('stasiun asal : ',asal[12:])
    globals()['multiplier_'+asal[12:]]=[]
    
    for j in range (1,len(dataset_rainfall.columns)):
        if (i!=j):
            tujuan=dataset_rainfall.columns[j]
            print ('stasiun tujuan : ',tujuan[12:])
            
            # The commonly used p is 2
            p=2 
            
            # get the d
            try:
                d=df_jarak.loc[(df_jarak['Stasiun Asal'] == asal) & (df_jarak['Stasiun Acuan'] == tujuan)]['Jarak'].values[0]
            except:
                d=df_jarak.loc[(df_jarak['Stasiun Asal'] == tujuan) & (df_jarak['Stasiun Acuan'] == asal)]['Jarak'].values[0]
            
            # calculate the weight
            weight_idw=(1/d)**p
            
            globals()['multiplier_'+asal[12:]].append([asal,tujuan,weight_idw])
    
    # change the list of multiplier into dataframe
    globals()['multiplier_'+asal[12:]]=pd.DataFrame(globals()['multiplier_'+asal[12:]], columns =['Asal', 'Acuan','Weight IDW'])    
    print ('')
```

## Calculating IDW rainfall

```python
globals()['multiplier_'+asal[12:]]
```

Pada section ini akan dibuat variabel per masing-masing stasiun untuk menyimpan curah hujan pengganti yang menggunakan metode IDW. Akan dibuat variabel `IDW_(nama stasiun)` dengan looping menggunakan `globals()['IDW_(nama stasiun)']`.

Pada section sebelumnya kita telah mendapatkan dataframe dengan penulisan nama `multiplier_(nama stasiun)`, faktor pengali tersebut akan dikalikan dengan nilai CH per stasiun acuan dan didapatkan hujan IDW per nama stasiun.

#### Replacing Missing Value with IDW

```python
list_sta=dataset_rainfall.columns[1:].values

for sta in list_sta:
    for i in range (len(globals()['multiplier_'+sta[12:]])):
        idw=0
        
        sta_acuan=globals()['multiplier_'+sta[12:]]['Acuan'][i]
        weight_idw=globals()['multiplier_'+sta[12:]]['Weight IDW'][i]
        ch_acuan=dataset_rainfall[sta_acuan]
        
        idw=idw+(ch_acuan.mul(weight_idw, fill_value=0))
    globals()['IDW_'+sta[12:]]=idw
    dataset_rainfall[sta][dataset_rainfall[sta].isnull()]=globals()['IDW_'+sta[12:]]
```
