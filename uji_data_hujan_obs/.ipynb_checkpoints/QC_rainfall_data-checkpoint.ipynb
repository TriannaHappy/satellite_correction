### Import Libraries

```python
import pandas as pd
import glob
import numpy as np
import seaborn as sns
```

### Data Loading

```python
# list all the rainfall station data
list_st=glob.glob('D:\\Happy\\RTOW\\Pemali Juana\\koreksi gpm\\CH koreksi\\Dipakai\\utk koreksi fix\\*.xls')
list_st[0][74:-4]
```

```python
# load the dataset
dataset=pd.read_excel("D:/Happy/RTOW/Pemali Juana/koreksi gpm/CH koreksi/CHpemali_all.xlsx")
dataset.info()
```

We have 60 existed stations.

```python
# create the dataframe
df=pd.DataFrame({"date":pd.date_range(start="2011-01-01",end="2020-12-31")})
df.head()
```

### Handling Missing Value

```python
# append station name
for i in list_st:
    df[i[74:-4]]=dataset.loc[:,i[74:-4]]

# 
df=df.replace('-', np.nan)
df.info()
```

```python
df.loc[:,"J-SE.161-Pancur"]=df.loc[:,"J-SE.161-Pancur"].replace('7`',7)
df.info()
```

```python
df2=df.copy()
for i in range (1,df2.shape[0]):
    df2.iloc[i,1:]=df2.iloc[i-1,1:]+df2.iloc[i,1:]
df2
```

```python
df2["Month"]=df2["date"].dt.month
df2_month=df2.groupby(["Month"]).mean()
df2_month=df2_month.reset_index()
df2_month
```

```python
df2_month.plot(x="Month", y=df2_month.columns[1:].values,
        kind="line", figsize=(10, 8))
```

```python
df3["Month"]=df3["date"].dt.month
df3_month=df3.groupby(["Month"]).mean()
df3_month=df3_month.reset_index()
df3_month
```
